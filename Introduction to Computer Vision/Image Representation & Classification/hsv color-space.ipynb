```python
#import the libraries

import cv2 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#defining the paths for the image 

img_path = "./images/"
img = "water_balloons.jpg"

# reading the image
img = cv2.imread(img_path+img)
```

```python
# RGB color space 
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

red_channel = rgb_img[:,:,0]
green_channel = rgb_img[:,:,1]
blue_channel = rgb_img[:,:,2]

# ploting channels on gray scale
f,(ax0,ax1,ax2,ax3) = plt.subplots(1,4,figsize=(20,30))
ax0.set_title("RGB Image")
ax0.imshow(rgb_img)
ax1.set_title("red")
ax1.imshow(red_channel,cmap='gray')
ax2.set_title("blue")
ax2.imshow(blue_channel,cmap='gray')
ax3.set_title("green")
ax3.imshow(green_channel,cmap='gray')
```

```python
# creating the mask
pink_low = np.array([180 , 0 , 100])
pink_high = np.array([255 , 255 ,230])
rgb_mask = cv2.inRange(rgb_img,pink_low,pink_high)

# applying the mask

img_copy1 = np.copy(rgb_img)
img_copy1[rgb_mask==0] = [0,0,0]

# plotting the masked image
plt.imshow(img_copy1)
```

```python
# HSV Color space
hsv_img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

h = hsv_img[:,:,0]
s = hsv_img[:,:,1]
v = hsv_img[:,:,2]

# ploting channels on gray scale
f,(ax0,ax1,ax2,ax3) = plt.subplots(1,4,figsize=(20,30))
ax0.set_title("HSV Image")
ax0.imshow(hsv_img)
ax1.set_title("Hue")
ax1.imshow(h,cmap='gray')
ax2.set_title("Saturation")
ax2.imshow(s,cmap='gray')
ax3.set_title("value")
ax3.imshow(v,cmap='gray')
```

```python
# creating the mask
low= np.array([160,0,0]) 
high = np.array([180,255,255])
hsv_mask = cv2.inRange(hsv_img,low,high)

# applying the mask
img_copy1 = np.copy(rgb_img)
img_copy1[hsv_mask==0] = [0,0,0]

#plotting the masked image
plt.imshow(img_copy1)
```
